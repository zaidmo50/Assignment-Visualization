```python
#4. Gen 3D surface plot to visualize z=sin((x^2+y^2)^1/2)
x = np.linspace(-5, 5, 10)
y = np.linspace(-5, 5, 10)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df8 = pd.DataFrame(data)
```

```python
fig = go.Figure(data=[go.Surface(z=z, x=x, y=y)])
fig.show()
```

```python
#1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.
```

```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
```

```python
from bokeh.plotting import figure,show
x = np.linspace(0, 10, 100)
y = np.sin(x)
p=figure()
p.line(x,y,line_width=2)
show(p)
```

```python
#2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.
```

```python
x=np.random.randint(0,100,20)
y=np.random.randint(0,100,20)
sizes = np.random.randint(10, 50, size=20)
colors = np.random.choice(['red', 'green', 'blue', 'orange', 'purple'], size=20)
p=figure()
p.scatter(x,y,size=sizes,marker='circle',color=colors)
show(p)
```

```python
#3.Generate a Bokeh bar chart representing the counts of different fruits using the following dataset
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]
```

```python
p=figure(x_range=fruits,title='Fruits Count')
p.xaxis.axis_label='Name'
p.yaxis.axis_label='Count'

p.vbar(x=fruits,top=counts)
show(p)
```

```python
#4.Create a histogram
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)
p=figure(title="Histogram")
p.quad(top=hist,bottom=0,left=edges[:-1],right=edges[1:])
show(p)
```

```python
#5. Create a bokeh heatmap using
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)
```

```python
from bokeh.palettes import Viridis256
from bokeh.transform import linear_cmap

p = figure(title="Heatmap",x_range=(0, 1), y_range=(0, 1))
# Flatten the arrays for Bokeh
data = {
    'x': xx.flatten(),
    'y': yy.flatten(),
    'value': data_heatmap.flatten()
}
color_mapper = linear_cmap(field_name='value', palette=Viridis256, low=min(data['value']), high=max(data['value']))

p.rect(x='x',y='y',source=data,width=0.1, height=0.1,fill_color=color_mapper)
show(p)
```
